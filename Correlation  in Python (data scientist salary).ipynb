```python
#sorting data 
df.sort_values(by=['Avg Salary(K)'], inplace=False, ascending=False)
df.head()
```

```python
#coorelating all numeric columns
df.corr(method ='spearman').head(5)
```

```python
#factorising non numeric columns
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Correlation Matrix between all numeric columns

correlation_matrix_pearson=df_numerized.corr(method='pearson')
correlation_matrix_pearson.head()
```

```python
correlation_matrix_kendall=df_numerized.corr(method='kendall')
correlation_matrix_kendall.head()
```

```python
correlation_matrix_spearman=df_numerized.corr(method='spearman')
correlation_matrix_spearman.head()
```

```python
#continuing with the spearman method
correlation_matrix_spearman_sort = correlation_matrix_spearman.unstack()
correlation_matrix_spearman_sort

```

```python
correlation_matrix_spearman_sort = correlation_matrix_spearman.unstack()
print(correlation_matrix_spearman_sort[798:840])
```

```python
# We can now take a look at the ones that have a high correlation (> 0.2)
strong_pairs = (correlation_matrix_spearman_sort[798:840])[abs((correlation_matrix_spearman_sort[798:840])) > 0.1].sort_values(inplace=False, ascending=False, kind="quicksort") 
strong_pairs

```
